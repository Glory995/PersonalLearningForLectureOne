#  <center> Programming Languages </center>
## Low vs. High level
- ### High level languages are more Understandable by Humans
    #### Examples: C/C++, Java, Python, Go, TypeScript, Julia
- ### Low level languages are closer to machine (zeros and ones)
    #### x86 Assembly, ARM Assembly

## Interpreted vs. Compiled
- ### Compiled languages are first converted to binary executables and then run
    #### Examples: C/C++, Java, C#, Kotlin, Swift
- ### Interpreted languages are converted to binary line by line and executed
    #### Examples: JavaScript, Python



#  <center> Programming Languages </center>
## General vs. Targeted
- ### Targeted languages are made for a specific kind of applications
    #### Examples: SQL for databases, HTML for web design, etc.
- ### General puprpose languages can be used for a wide range of applications
    #### Examples: C/C++, Python, Java, Swift, JavaScript

# <center> Conditionals </center> 
- ## Write a program that among three variables x, y, and z, prints the largest odd number and if none of them is odd then prints the smallest number

```python
x, y, z = 5, 3, 8
output = min(x,y,z)
if x%2 != 0:
    output = x
if y%2 != 0 and y>output:
    output = y
if z%2 != 0 and z>output:
    output = z
print(output)
```

# <center> Conditional Expressions </center> 

- ## A conditional expression is a python expression the value of which depends on a condition
### expression1 _if_ condition _else_ expression2
- ## Example: Finding the minimum of two numbers:


```python
x, y = 5, 3

if x<y:
    minimum = x
else:
    minimum = y
print("Minimum using normal if statement:", minimum)

minimum = x if x < y else y

print("Minimum using conditional expression:", minimum)
```

# <center> String Slicing </center> 
## Slicing is a common operation in Python. You can take a slice of a string by specifying the start (inclusive) and end (exclusive) indices:

```python
s = "beta"
s[1:3] # start index is 1 and end index is 3
```

```python
num="01234567"
num[0:len(num):2] # start:0, stop: len(num), step:2
```

```python
num[::2] # not specifying start makes it 0, not specfying stop makes it the length of the string
```
